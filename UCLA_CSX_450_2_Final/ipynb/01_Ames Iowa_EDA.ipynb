Looks like in which month the house was sold didn't affect the sale price, but the external quality and house style affected the sale price. The houses with excellent external quality were sold in a higher price. House with 2.5 finished stories were sold in a higher price. More sold houses are in split foyer style. 

```r
multiplot(hist_with_kde_numerical_by_category(SalePrice,MoSold),
          hist_with_kde_numerical_by_category(SalePrice,ExterQual),
          hist_with_kde_numerical_by_category(SalePrice,HouseStyle),
          cols = 2)
```

```r
meansd = function(x) c(mean=mean(x), sd=sd(x))
by(SalePrice, MoSold, FUN=meansd)
```

```r
oneway.test(SalePrice ~ MoSold)
```

```r
oneway.test(SalePrice ~ ExterQual)
```
