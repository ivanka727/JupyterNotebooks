```r
Boston <- read.table('Boston.csv', sep=",", header = T)
```

```r
Boston$X <- NULL
```

```r
head(Boston)
```

```r
dim(Boston); str(Boston)
```

```r
summary(Boston)
```

```r
Boston$chas = factor(Boston$chas)
Boston$medv <- NULL
Boston_features = Filter(is.numeric, Boston)
```

```r
sum_vals = data.frame(feature=colnames(Boston_features))
sum_vals['mean_'] = sapply(Boston_features, mean)
sum_vals
```

```r
library(dplyr, warn.conflicts = FALSE)
samples = list()
for (i in 1:10) {
    samples[[i]] = sample_n(Boston_features, 5)
}
```

```r
for (i in 1:10) {
    sum_vals[paste('mean_', i)] = sapply(samples[[i]], mean)
}
```

```r
sum_vals
```

```r
library(reshape2)
library(ggplot2)

ggplot(melt(sum_vals), aes(x = feature, y = value, fill = variable)) + 
    geom_bar(stat = "identity", width=0.5, position = "dodge")
```

### 5 samples have been selected, and we can see there is a big deviation with the actual mean. So we need to select more samples.

```r
library(dplyr, warn.conflicts = FALSE)
samples = list()
for (i in 1:10) {
    samples[[i]] = sample_n(Boston_features, 50)
}
```

```r
for (i in 1:10) {
    sum_vals[paste('mean_', i)] = sapply(samples[[i]], mean)
}
```

```r
library(reshape2)
library(ggplot2)

ggplot(melt(sum_vals), aes(x = feature, y = value, fill = variable)) + 
    geom_bar(stat = "identity", width=0.5, position = "dodge")
```

### 50 samples have been selected, we can see the deviation became smaller, but the sample size is still not good to use.
